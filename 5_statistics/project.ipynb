```python
from matplotlib import pyplot as plt
import numpy as np

# Salary data
salaries = np.array([3000, 3200, 3500, 4000, 4200, 5000, 12000])

# Statistics
mean_salary = salaries.mean()
median_salary = np.median(salaries)
std_salary = salaries.std()

# Plot
plt.figure()
plt.barh(range(len(salaries)), salaries)

# Mean & median lines
plt.axvline(mean_salary, linestyle='--', label=f"Mean ({mean_salary:.0f})")
plt.axvline(median_salary, linestyle='-', label=f"Median ({median_salary:.0f})")

# Labels / title
plt.yticks(range(len(salaries)), [f"Emp {i+1}" for i in range(len(salaries))])
plt.xlabel("Salary ($)")
plt.title("Salary Distribution (Horizontal)")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------
# Salary Data
# ------------------------
salaries = np.array([3000, 3200, 3500, 4000, 4200, 5000, 12000])

# ------------------------
# Statistical Calculations
# ------------------------
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)
std_salary = np.std(salaries)
var_salary = np.var(salaries)
min_salary = np.min(salaries)
max_salary = np.max(salaries)
argmin_salary = np.argmin(salaries)
argmax_salary = np.argmax(salaries)
p25 = np.percentile(salaries, 25)
p75 = np.percentile(salaries, 75)

# Print calculations
print(f"Mean: {mean_salary}")
print(f"Median: {median_salary}")
print(f"Std Dev: {std_salary}")
print(f"Variance: {var_salary}")
print(f"Min: {min_salary}, Max: {max_salary}")
print(f"Index Min: {argmin_salary}, Index Max: {argmax_salary}")
print(f"25th Percentile: {p25}, 75th Percentile: {p75}")

# ------------------------
# Visualization with Matplotlib
# ------------------------
plt.figure(figsize=(10, 6))
plt.barh(range(len(salaries)), salaries, color='skyblue', edgecolor='black')
plt.yticks(range(len(salaries)), [f"Emp {i+1}" for i in range(len(salaries))])
plt.xlabel("Salary ($)")
plt.title("Salary Distribution with Statistical Indicators")

# Mean, Median, and Std Dev lines
plt.axvline(mean_salary, color='orange', linestyle='--', linewidth=2, label=f"Mean = {mean_salary:.0f}")
plt.axvline(median_salary, color='green', linestyle='-', linewidth=2, label=f"Median = {median_salary:.0f}")

# Shaded region for standard deviation range
plt.axvspan(mean_salary - std_salary, mean_salary + std_salary, color='orange', alpha=0.15, label="±1 Std Dev")

# Mark min and max points
plt.scatter(min_salary, argmin_salary, color='red', label=f"Min = {min_salary}")
plt.scatter(max_salary, argmax_salary, color='purple', label=f"Max = {max_salary}")

# Percentile lines
plt.axvline(p25, color='gray', linestyle=':', linewidth=2, label=f"25th = {p25:.0f}")
plt.axvline(p75, color='gray', linestyle=':', linewidth=2, label=f"75th = {p75:.0f}")

plt.legend()
plt.show()
```
