```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 1. Load the image
img = Image.open('/Users/newuser/Desktop/gen-ai/gen_ai_courses/week_5/8_image_processing/dog.png')  # put any small image in the same folder
img_np = np.array(img)

print("Original Shape:", img_np.shape)  # (Height, Width, Channels)
```

```python
# 2. Convert to grayscale
#  WEIGHTS: R:0.299, G:0.587, B:0.114 (standard luminance conversion)
gray = 0.299 * img_np[:,:,0] + 0.587 * img_np[:,:,1] + 0.114 * img_np[:,:,2]
gray = gray.astype(np.uint8)
```

```python
# 3. Apply a simple edge-detection filter using NumPy convolution
# Sobel-like kernel
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])
```

```python
h, w = gray.shape
edge = np.zeros_like(gray)

for i in range(1, h-1):
    for j in range(1, w-1):
        region = gray[i-1:i+2, j-1:j+2]
        edge[i,j] = np.sum(region * kernel)
```

```python
# 4. Visualization
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img_np)
plt.axis('off')
```

```python
plt.subplot(1,3,2)
plt.title("Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(1,3,3)
plt.title("Edge Detection")
plt.imshow(edge, cmap='gray')
plt.axis('off')
```
