```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load face image
img = Image.open("/Users/newuser/Desktop/gen-ai/gen_ai_courses/week_5/8_image_processing/face.png")  # put any face image in same folder
img_np = np.array(img)
```

```python

# Convert to grayscale
gray = 0.299 * img_np[:,:,0] + 0.587 * img_np[:,:,1] + 0.114 * img_np[:,:,2]
gray = gray.astype(np.uint8)

# Edge detection kernel
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Convolution
h, w = gray.shape
edge = np.zeros_like(gray)

for i in range(1, h-1):
    for j in range(1, w-1):
        region = gray[i-1:i+2, j-1:j+2]
        edge[i,j] = np.sum(region * kernel)

# Show result
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original Face")
plt.imshow(img_np)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Edges (Face Shape)")
plt.imshow(edge, cmap='gray')
plt.axis('off')

plt.show()
```
